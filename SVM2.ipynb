```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(-10.0,10.0,100)
x
```

```python
y=np.sqrt(10**2-x**2)
```

```python
y
```

```python
x1=np.hstack([x,-x])
y1=np.hstack([y,-y])
```

```python
x1
```

```python
y1
```

```python
x2=np.linspace(-5.0,5.0,100)
y2=np.sqrt(5**2-x2**2)
x2=np.hstack([x2,-x2])
y2=np.hstack([y2,-y2])
```

```python
plt.scatter(y1,x1)
plt.scatter(y2,x2)
plt.show()
```

```python
df1=pd.DataFrame(np.vstack([y1,x1]).T,columns=['x1','x2'])
df1['y']=0
```

```python
df1.head()
```

```python
df2=pd.DataFrame(np.vstack([y2,x2]).T,columns=['x1','x2'])
df2['y']=1
df2.head()
```

```python
df=pd.concat([df1,df2])
df.head()
```

```python
df['y'].unique()
```

```python
x=df.iloc[:,:2]
x.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.svm import SVC,SVR
```

```python
classifier=SVC(kernel='rbf')
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
classifier=SVC(kernel='poly')
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
