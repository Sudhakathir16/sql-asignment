```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

plt.plot(x, y, color='teal', linewidth=3, marker='*',
         markersize=15, linestyle='--')

plt.title("Customizing Line Chart")

plt.ylabel('Y-Axis')

plt.xlabel('X-Axis')

plt.show()
```

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
sns.load_dataset('tips')
```

```python
tips = sns.load_dataset('tips')
sns.lineplot(x='total_bill', y='smoker',linestyle="--", data=tips)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

```python
sns.countplot(x='day', data=tips)
```

```python
sns.histplot(x='total_bill', data=tips)
```

```python
sns.pairplot(tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
numeric_tips = tips.select_dtypes(include=['number'])
sns.heatmap(numeric_tips.corr(), annot=True)
```
