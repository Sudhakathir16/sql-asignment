```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
data = {
    'Age': [22, 25, 47, 52, 46, 56, 55],
    'Income': ['High', 'High', 'Medium', 'Low', 'Low', 'Medium', 'High'],
    'Buys Computer': ['No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No']
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Convert categorical variables to numerical
df['Income'] = df['Income'].map({'High': 1, 'Medium': 2, 'Low': 3})
df['Buys Computer'] = df['Buys Computer'].map({'No': 0, 'Yes': 1})
```

```python
df['Income']
```

```python
df['Buys Computer']
```

```python
# Features and target variable
X = df[['Age', 'Income']]
y = df['Buys Computer']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
cla=DecisionTreeClassifier()
cla.fit(X_train,y_train)
```

```python
cla=DecisionTreeClassifier()
cla.fit(X_test,y_test)
```

```python
# Visualize the decision tree
plt.figure(figsize=(5, 2))
tree.plot_tree(cla, feature_names=['Age', 'Income'], class_names=['No', 'Yes'], filled=True)
plt.show()
```

```python
# Make predictions
predictions = cla.predict(X_test)
print("Predictions:", predictions)
```

```python
accuracy = cla.score(X_test, y_test)
print("Accuracy:", accuracy)
```
